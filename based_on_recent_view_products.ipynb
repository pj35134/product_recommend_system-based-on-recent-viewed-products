```
#customer or user id of all the customers who viewed at least one of the products
customer_viewed = m[m.event_type== 'view'].user_id.unique()
#for customer in customer_purchased:
customer_viewed[:100]
```

```

for x in customer_viewed:
  viewed_items=[]

  #get productid of all the products that an individual customer has viewed sorted with timestamp
  print('\n\n\n\n user_id',x)
  viewed_items=m.loc[(m.user_id ==  x) & (m.event_type=='view')].sort_values('event_time',ascending=False).product_id.values
  #print('\n viewed items',viewed_items)
  if len(viewed_items)>0:
    k=viewed_items[0]
  #print("first viewed items",k)

  #create new dataframe to add recently viewed items 

  df = pd.DataFrame(viewed_items, columns =['Recent']) 

  #count the duplicates or frequency and add it to dataframe

  mydf = df.groupby(['Recent'],sort=False).size().reset_index()
  mydf.rename(columns = {0: 'frequency'}, inplace = True)
  
  #calculate maximum frequency 

  maxValue = mydf.max()['frequency']
  
  #for each frequency divide it by max freq and add it to dataframe

  c=[]
  for i in mydf['frequency']:
    c.append(i/maxValue)

  mydf['fnorm']=c

  #total length of a column named Recent
  total_rows = len(mydf['Recent'])
  #print('total_rows',total_rows)

  # To get the product that has been mostly visited
  newe=[]
  most_view_product=[]

  #for dataframe having more than 2 rows(viewed items) for each individual user
  if total_rows>=2:
    # normalizing the Recent column values from 1 to 0 i.e the top most or recent part is given higher priority(1)
    for j in reversed(range(total_rows)):
      #print('j is ',j/(total_rows-1))
      newe.append(j/(total_rows-1))
    #add new normalized values in dataframe
    mydf['normalized_recency']=newe

    #getting the total value
    mydf['Value'] = mydf.fnorm * 0.25 +mydf.normalized_recency*0.75
    #sorting the dataframe in descending order with the highest total value at the top 
    mydf.sort_values(by=['Value'], inplace=True,ascending=False)
    # add new recent values after sorting to a list
    for finalrecommend in mydf['Recent']:
      # presenting products on the basis of total value i.e keeping recent timestamp and frequency on priority
      most_view_product.append(finalrecommend)
      
  elif total_rows<= 1:
    for finalrecommend in mydf['Recent']:
      #print('if row is less than or equal to one')
      most_view_product.append(finalrecommend)
      
  print('\n final list of most viewed product based on recency and frequency',most_view_product)
  # suggesting similar brand and category products based on the final products obtained from the above(most_view_product)
  suggest_products=[]
  for similarproduct in most_view_product:
    #get the brand name and category name for each product 
    similar_products_brand=[]
    similar_products_category=[]
    similar_prodcts_name=[]
    similar_products_brand=m.loc[m.product_id==similarproduct].brand.unique()
    exact_brand=similar_products_brand[0]
    #print('\n similar_brand_products',exact_brand)
    similar_products_category=m.loc[m.product_id==similarproduct].main_category_list.unique()
    exact_category=similar_products_category[0]
    #print('\n similar_main_category_products',exact_category)
    similar_products_name=m.loc[m.product_id==similarproduct].Product.unique()
    exact_product_name=similar_products_name[0]
    #print('\n similar_products_name',exact_product_name)
    #get similar products based on the brand and category and recommend it
    suggest_products.append(list(m.loc[(m.brand==exact_brand) & (m.Product==exact_product_name)].product_id.values[:4]))
    #print('\n suggested products\n',suggest_products)
    finallist=[]
    finallist=suggest_products
    
    
    #print('\n set products',set(suggest_products))



  print('suggested products',finallist)
  
  

```
